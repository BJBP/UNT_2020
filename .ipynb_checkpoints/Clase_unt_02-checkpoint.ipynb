# Scikit-Learn
Biblioteca completa para analisis de datos, minería de datos y aprendizaje de máquina (http://scikit-learn.org/stable).

image.png

```python
# bilioteca de graficos
import matplotlib.pyplot as plt
# modelo de regresion lineal
from sklearn import linear_model
# media de los quadrados de las diferencias. Medida de error de un modelo
# r2-score: el coeficiente de determinación del  modelo de regresion lineal
from sklearn.metrics import mean_squared_error, r2_score
# particiona en grupos de entrenamiento y teste
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
# cargando el dataset iris
from sklearn import datasets
iris = datasets.load_iris()
```

```python
# covertindo para dataframe
iris = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
columns = iris['feature_names'] + ['species'])
```

```python
# visualizando os dados (os 5 primeiros)
iris.head()
```

```python
# agrupación por grupos
iris.groupby(by = "species").mean().plot(kind="bar")
plt.title('Clases vs Medidas')
plt.ylabel('Média de la medidas(cm)')
plt.xticks(rotation=0)
plt.grid(True)
plt.legend(loc="upper left", bbox_to_anchor=(1,1))
plt.show()
```
